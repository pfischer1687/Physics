```python
# Let us begin by importing the necessary libraries and/or functions.
from math import sqrt, ceil, floor
from scipy.special import binom
import matplotlib.pyplot as plt
from IPython.display import display, Latex
```

```python
# Let us define functions to calculate the total values for S_N and the standard deviation from the mean
# for an element N of the set of nonnegative integers.

# total values for S_N
def totValsSN(n):
    return 2**n

# standard deviation from the mean
def sigmaN(n):
    return sqrt(n)

def fracWInSigma(n):
    stDev = sigmaN(n)
    xAxis = list(range(n, -(n+1), -2))
    howManyWays = []
    for i in range(0, n+1):
        howManyWays.append(binom(n, abs(i)))

    xValsWInStDev = [i for i in xAxis if (i >= -stDev) & (i <= stDev)]
    yvalsWInStDev = [howManyWays[i] for i in range(floor((len(howManyWays)-len(xValsWInStDev))/2),floor((len(howManyWays)+len(xValsWInStDev))/2))]
    frac = sum(yvalsWInStDev) / totValsSN(n)

    display(Latex(f'The fraction of the total values for $S_N$ that are within $\pm\sigma_N$ of the mean: {frac}'))

    plt.plot(xAxis,howManyWays,'ro')
    plt.axvline(x=-stDev)
    plt.axvline(x=stDev)
    plt.axis([-(n+1), n+1, -n, binom(n,n/2)+n])
    plt.xticks(xAxis)
    plt.xlabel('$S_N$')
    plt.ylabel('# ways to sum to $S_N$')
    plt.title('# ways to sum to $S_N$ for N = %d' % n)
    plt.text(-n, binom(n,n/2)-n, '$\sigma_N = $%.2f' % stDev)
    plt.annotate('$+\sigma_N$', xy=(stDev, 3/4*binom(n,n/2)), xytext=(1.5*stDev, 13/16*binom(n,n/2)),
                arrowprops=dict(facecolor='black', shrink=0.05))
    plt.annotate('$-\sigma_N$', xy=(-stDev, 3/4*binom(n,n/2)), xytext=(-1.9*stDev, 13/16*binom(n,n/2)),
                arrowprops=dict(facecolor='black', shrink=0.05))
    plt.show()
```

```python
for n in [4, 5, 6, 10, 20]:
    fracWInSigma(n)
```
